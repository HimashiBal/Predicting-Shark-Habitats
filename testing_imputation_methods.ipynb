```python
sss=pd.read_csv('/content/drive/MyDrive/SHARK DATA/cleaned data/sss_0_8214.csv')
sst=pd.read_csv('/content/drive/MyDrive/SHARK DATA/cleaned data/sst_0_8214.csv')
chl=pd.read_csv('/content/drive/MyDrive/SHARK DATA/cleaned data/chl_0_8214.csv')
z=pd.read_csv('/content/drive/MyDrive/SHARK DATA/cleaned data/z_0_8214.csv')
dtc=pd.read_csv('/content/drive/MyDrive/SHARK DATA/cleaned data/dtc_0_8214.csv')
```

```python
sss_interp = sss.interpolate(method ='linear', limit_direction ='forward')
sst_interp = sst.interpolate(method ='linear', limit_direction ='forward')
chl_interp = chl.interpolate(method ='linear', limit_direction ='forward')
z_interp = z.interpolate(method ='linear', limit_direction ='forward')
dtc_interp = dtc.interpolate(method ='linear', limit_direction ='forward')
```

```python
df_mpm=pd.read_csv("/content/drive/MyDrive/SHARK DATA/cleaned data/MPM_Data_correct.csv")
```

```python
df_mpm.head()
```

```python

sss_col = sss_interp["PSU"]
merged_data = pd.concat([df_mpm, sss_col.rename("sss")], axis=1)
chl_col = chl_interp["mg m-3"]
merged_data = pd.concat([merged_data, chl_col.rename("chl")], axis=1)
z_col = z_interp['m']
merged_data = pd.concat([merged_data, z_col.rename("z")], axis=1)
sst_col = sst_interp['C']
merged_data = pd.concat([merged_data, sst_col.rename("sst")], axis=1)
dtc_col = dtc_interp['km']
merged_data = pd.concat([merged_data, dtc_col.rename("dtc")], axis=1)

merged_data.head()
# merged_data.to_csv('cleaned_data.csv')
```

```python
merged_data = merged_data.drop(['x', 'y', 'x.se', 'y.se', 'logit_g', 'logit_g.se'], axis=1)
```
